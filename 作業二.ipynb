```python
import numpy as np
import matplotlib.pyplot as plt
K1=8700 
K2=9100 
K3=9500
Interval=800
Premium_Call1=80
Premium_Call2=60
Premium_Call3=40
ST=np.arange(K2-Interval,K2+Interval)
Payoff_LongCall1=np.maximum(ST-K1,0)-Premium_Call1
Payoff_ShortCall2=np.maximum(ST-K2,0)+Premium_Call2
Payoff_LongCall3=np.maximum(ST-K3,0)-Premium_Call3
plt.plot(ST,Payoff_LongCall1-2*Payoff_ShortCall2+Payoff_LongCall3)
plt.show()
```
